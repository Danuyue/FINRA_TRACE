```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')

from pathlib import Path
```

```python
root_folder = Path('../Data/')
file_name = 'f99999c0cc149c54.csv'
file_path = root_folder / file_name

field_of_interest = ['cusip_id', 'bond_sym_id', 'company_symbol', 'trd_exctn_dt',
       'trd_exctn_tm', 'trd_rpt_dt', 'trd_rpt_tm','entrd_vol_qt', 'rptd_pr',
       'rpt_side_cd', 'sell_cmsn_rt', 
        'cntra_mp_id', 'trdg_mkt_cd',
       'SUB_PRDCT', 'STLMNT_DT']

dtype={'cusip_id': str , 'bond_sym_id': str, 'company_symbol': str, 'cusip_id': str,
      'entrd_vol_qt' : np.int , 'rptd_pr' : np.float, 'rpt_side_cd':str, 'sell_cmsn_rt':np.float32,
      'cntra_mp_id':str, 'trdg_mkt_cd':str, 'SUB_PRDCT':str}

parse_dates = [['trd_exctn_dt','trd_exctn_tm'],['trd_rpt_dt','trd_rpt_tm'],'STLMNT_DT']

data = pd.read_csv(file_path,usecols=field_of_interest,parse_dates=parse_dates,
                   infer_datetime_format=True)
```
